Transformação dos dados 





```python
#Sustituindo da coluna Var%: As virgula dos dados por ponto, retirando a porcentagem e passando os dados para float e dividindo por 100
df2['Var%'] = df2['Var%'].str.replace(",", ".").str.replace("%", "").astype(float)/100
```

```python
#Sustituindo da coluna Vol.: K e M pelos respectivos zeros correspondentes, retirando a virgula e passando os dados para inteiro
df2['Vol.'] = df2['Vol.'].str.replace("K", "0").str.replace("M", "0000").str.replace(",", "").astype(int)
```

```python
#Sustituindo da coluna Data: O formato da data de dia, mês e ano para ano, mês e dia e passando os dados para datetime
df2['Data'] = pd.to_datetime(df['Data'], errors='coerce')
df2['Data'] = df2['Data'].dt.strftime('%Y-%m-%d').astype('datetime64[ns]')
```

```python
from pyspark.sql import SparkSession, Window
spark = SparkSession.builder.appName('Teste').getOrCreate()

import pyspark.sql.functions as F
from pyspark.sql.functions import to_timestamp, to_date

from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType,BooleanType,DateType,TimestampType
```

```python
#Criando um esquema para o dataframe
customSchema = StructType([
   StructField("_c0", IntegerType(),True),                        
   StructField("Data", DateType(),True),
   StructField("Ultimo", DoubleType(),True),
   StructField("Abertura", DoubleType(),True),
   StructField("Maxima", DoubleType(),True),
   StructField("Minima", DoubleType(),True),
   StructField("Volume", IntegerType(),True), 
   StructField("Variacao", DoubleType(),True),

])

schema = customSchema
```

```python
#Importando o csv e criando um novo dataframe
df3 = spark.read.option("header","true").schema(schema).csv("/content/Dados_Ibovespa_tratado.csv")
```

```python
df3.printSchema()
```

```python
df3.show(5)
```

Queries com PySpark

```python
#Criando uma visualização com um ranking do Volume com o maior valor negociado no dia
df_rkn = df3.groupBy(F.col('Volume'), F.col('Abertura'), F.col('Ultimo'), F.col('Variacao')).agg(
    F.max(F.col('Maxima')).alias('Negociacao_Maxima')
)

w1 = Window.orderBy(F.col('Negociacao_Maxima').desc())

df_rkn = (df_rkn.select(F.col('Abertura'), F.col('Ultimo'), F.col('Variacao'), F.col('Volume'), F.col('Negociacao_Maxima'), ).withColumn("Ranking", F.row_number().over(w1))) 
df_rkn.show(10)
```

```python
#Criando uma visualização com um ranking do Volume com o menor valor negociado no dia
df_rkn_min = df3.groupBy(F.col('Volume'), F.col('Abertura'), F.col('Ultimo'), F.col('Variacao')).agg(
    F.min(F.col('Minima')).alias('Negociacao_Minima')
)

w1 = Window.orderBy(F.col('Negociacao_Minima').asc())

df_rkn_min = (df_rkn_min.select(F.col('Abertura'), F.col('Ultimo'), F.col('Variacao'), F.col('Volume'), F.col('Negociacao_Minima'), ).withColumn("Ranking", F.row_number().over(w1))) 
df_rkn_min.show(10)
```

SparkSQL

```python
#Criação de uma visão temporária local 
df3.createOrReplaceTempView("Tabela_Ibovespa")
```

```python
#Retornar todos os dados da Tabela_Ibovespa
spark.sql("SELECT * FROM Tabela_Ibovespa").show(10)
```

```python
#Retornar os valores máximos de negociação acima de 100.000 decrecente 
spark.sql("SELECT * FROM Tabela_Ibovespa WHERE Maxima > 100.000 ORDER BY Maxima DESC").show()
```

```python
#Retornar os valores mínimos de negociação menor de 100.000 por ordem ascendente
spark.sql("SELECT * FROM Tabela_Ibovespa WHERE Minima < 100.000 ORDER BY Minima ASC").show()
```

```python
#Mostrar os dados do Ibosbespa no período entre 01-01-2020 e 31-01-2020
spark.sql("SELECT * FROM Tabela_Ibovespa WHERE Data BETWEEN '2020-01-01' AND '2020-12-31' ORDER BY Volume DESC LIMIT 10").show()
```

```python
#Mostrar os dados do Ibosbespa no período entre 01-01-2021 e 31-12-2021
spark.sql("SELECT * FROM Tabela_Ibovespa WHERE Data BETWEEN '2021-01-01' AND '2021-12-31' ORDER BY Volume DESC LIMIT 10").show()
```

```python
#Mostrar os dados do Ibosbespa no período entre 01-01-2022 e 31-12-2022
spark.sql("SELECT * FROM Tabela_Ibovespa WHERE Data BETWEEN '2022-01-01' AND '2022-03-08' ORDER BY Volume DESC LIMIT 10").show()
```

```python
#Filtrar o maior e menor volume negociação do dia 
spark.sql("SELECT MAX (Volume) AS Maior_Volume_Negociacao, MIN(Volume) AS Menor_Volume_Negociado FROM Tabela_Ibovespa").show()
```
